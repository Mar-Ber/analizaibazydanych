## Zadanie 3

Zdefiniuj poniższą funkcję i sporządź jej wykres dla argumentów z danego przedziału:

$f(x)=x^2+5$

    1. $x>-1 oraz x<1$
    2. $x>-6 oraz x<6$
    3. $x>0 oraz x<5$

Wspierając się dokumentacją Matplotlib. Dodaj do wykresu etykiety osi, tytuły wykresów i legendy.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fun(x):
    y = x**2+5
    return y
```

```python
x1 = np.linspace(-1,1)
y1 = fun(x1)
plt.plot(x1,y1)
plt.grid()
plt.title("Wykres funkcji dla x1")
plt.xlabel("x1")
plt.ylabel("y1")
plt.show()
```

```python
x2 = np.linspace(-6,6)
y2 = fun(x2)
plt.plot(x2,y2)
plt.grid()
plt.title("Wykres funkcji dla x2")
plt.xlabel("x2")
plt.ylabel("y2")
plt.show()
```

```python
x3 = np.linspace(0,5)
y3 = fun(x3)
plt.plot(x3,y3)
plt.grid()
plt.title("Wykres funkcji dla x3")
plt.xlabel("x3")
plt.ylabel("y3")
plt.show()
```
